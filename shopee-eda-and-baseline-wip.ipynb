```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
dataset_path = Path('../input/shopee-product-matching')
```

Let's see what data is available:

```python
list(dataset_path.iterdir())
```

Let's check the `train.csv` file:

```python
train_df = pd.read_csv(dataset_path/'train.csv')
```

```python
train_df.head(10)
```

```python
print(f'There are {len(train_df)} postings and {len(train_df.label_group.unique())} label groups')
```

As you can see, we have a `posting_id`, a corresponding `image` (filename is provided), a `image_phash` (a perceptual hash), a `title` for the product posting, and the `label_group`. The purpose of this competition is to determine which postings are in the same `label_group`.

Let's also look at our `test.csv` and `sample_submission.csv`:

```python
test_df = pd.read_csv(dataset_path/'test.csv')
submission_df = pd.read_csv(dataset_path/'sample_submission.csv')
```

```python
test_df.head(10)
```

```python
submission_df.head(10)
```

We can see we are provided with similar information in `test.csv` except for the `label_group`. In the submission file, we don't predict which `label_group` the posting is part of, but rather construct our own `label_group`s by predicting what other postings match a given posting. Note that **all postings self match!**

Here is a simple baseline by finding which images have the same perceptual hash and providing them as matches:

```python
for i in tqdm(range(len(test_df))):
    submission_df.matches[i] = ' '.join(list(test_df.loc[test_df.image_phash == test_df.iloc[i].image_phash].posting_id))
```

```python
submission_df.to_csv('submission.csv', index=False)
```
